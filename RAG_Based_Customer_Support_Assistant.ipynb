```python
STEP 8: LLM Setup
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")

def simple_llm(prompt):
    result = generator(prompt, max_length=200, num_return_sequences=1)
    return result[0]['generated_text']
```

```python
STEP 9: Routing Logic
```

```python
def route_query(context, query):
    if not context or len(context.strip()) < 50:
        return "escalate"

    if "complaint" in query.lower():
        return "escalate"

    return "answer"
```

```python
STEP 10: HITL Function
```

```python
def human_escalation(query):
    print("\n⚠️ Escalated to Human Agent")
    print("Query:", query)
    return input("Enter human response: ")
```

```python
STEP 11: LangGraph Workflow
```

```python
from langgraph.graph import StateGraph
from typing import TypedDict

class GraphState(TypedDict):
    query: str
    context: str
    decision: str
    answer: str

def process_node(state):
    decision = route_query(state["context"], state["query"])
    return {**state, "decision": decision}

def answer_node(state):
    prompt = f"""
Context:
{state['context']}

Question:
{state['query']}

Answer clearly.
"""
    response = llm.invoke(prompt)
    return {**state, "answer": response.content}

def hitl_node(state):
    answer = human_escalation(state["query"])
    return {**state, "answer": answer}

builder = StateGraph(GraphState)

builder.add_node("process", process_node)
builder.add_node("answer", answer_node)
builder.add_node("hitl", hitl_node)

builder.set_entry_point("process")

builder.add_conditional_edges(
    "process",
    lambda state: state["decision"],
    {
        "answer": "answer",
        "escalate": "hitl"
    }
)

builder.set_finish_point("answer")
builder.set_finish_point("hitl")

graph = builder.compile()
```

```python
STEP 12: Query Execution
```

```python
def ask(query):
    docs = retriever.invoke(query)
    context = "\n".join([d.page_content for d in docs])

    decision = route_query(context, query)

    if decision == "escalate":
        answer = human_escalation(query)
    else:
        prompt = f"""
        Context:
        {context}

        Question:
        {query}

        Answer:
        """
        answer = simple_llm(prompt)

    print("\n💬 Answer:")
    print(answer)
```
